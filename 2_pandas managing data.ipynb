Pandas : Accessing Data

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\nikhi\Downloads\data_jobs.csv")
```

```python
# data cleanup

data['job_posted_date'] = pd.to_datetime(data['job_posted_date'])
```

```python
# ðŸ“Œ .loc â†’ Label
# ðŸ“Œ .iloc â†’ Integer (position)
```

```python
# now filling the NAN values with median values
```

```python
median_salary_year = data['salary_year_avg'].median()

median_salary_year
```

```python
median_salary_hour = data['salary_hour_avg'].median()

median_salary_hour
```

```python
# now we are making a new data frmae so that we can kept our original data frame untouched and do all the operation on the new one 
```

```python
df_filled = data

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
# now we are checking whethter the median values are filled in the places of NAN values or not

df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
# now we are dropping the duplicate values from out data.
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

print('Length of original data:        ',  len(df_filled))
print('Length of drop duplicates df:   ',  len(df_unique))
print('Rows Dropped:                   ',  len(df_filled)-len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

print('Length of original data:        ',  len(df_filled))
print('Length of drop duplicates df:   ',  len(df_unique))
print('Rows Dropped:                   ',  len(df_filled)-len(df_unique))
```
