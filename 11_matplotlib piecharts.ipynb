MATPLOTLIB: pie plots:

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import ast
#reading the csv file
data = pd.read_csv(r"C:\Users\nikhi\Downloads\data_jobs.csv")

# data cleanup
data['job_posted_date'] = pd.to_datetime(data['job_posted_date'])

data['job_skills'] = data['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
data['job_work_from_home'].value_counts().plot(kind='pie', startangle = 90, autopct='%1.1f%%')
plt.title('Work from home status')
plt.ylabel('')
plt.show()
```

```python
data['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job tidles in Dataset')
plt.ylabel('')
plt.show()
```

```python
data[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home' : 'work from home',
    'job_no_degree_mention' : 'job degree req.',
    'job_health_insurance' : 'Health insurance offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(data[column].value_counts(), startangle=90, autopct='%1.1f%%', labels = ['False','True'])
    ax[i].set_title(title)    
plt.show()
```
